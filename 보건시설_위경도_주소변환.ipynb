```python
cols=['위도','경도']
health['위도경도2']=health[cols].apply(lambda row:', '.join(row.values.astype(str)),axis=1)
health
```

```python
from geopy.geocoders import Nominatim

def geocoding_reverse(lat_lng_str): 
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    address = geolocoder.reverse(lat_lng_str)

    return address

#address = geocoding_reverse('37.58270882, 126.9982902')
#print(address)
```

```python
li=[]

for i in range(len(health['위도경도2'])):
    li2=[]
    address1= geocoding_reverse(health['위도경도2'][i])
    li2.append(address1)
    print(i,li2)
    li.append(li2)
#1시간 30분 걸림
```

```python
print(li)
```

```python
df1=pd.DataFrame(li)
df1
```

```python
df1.columns=["주소"]
df1
```
