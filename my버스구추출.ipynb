```python
from geopy.geocoders import Nominatim

def geocoding_reverse(lat_lng_str): 
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    address = geolocoder.reverse(lat_lng_str)

    return address

#address = geocoding_reverse('37.58270882, 126.9982902')
#print(address)
```

```python
li=[]

for i in range(len(kind['위도경도2'])):
    li2=[]
    address1= geocoding_reverse(kind['위도경도2'][i])
    li2.append(address1)
    print(i,li2)
    li.append(li2)
#1시간 30분 걸림
```
