```python
from dbfread import DBF
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
hgd = gpd.read_file('https://raw.githubusercontent.com/vuski/admdongkor/master/ver20211001/HangJeongDong_ver_20211001.geojson')
hgd
```

```python
hgd['adm_cd2'] = hgd['adm_cd2'].str[:-2]
hgd['adm_cd2'] = hgd['adm_cd2'].astype('int')
hgd = hgd.rename(columns={'adm_cd2': '행정동코드'})
hgd
```

```python
hgd = pd.concat([
    hgd,
    hgd['adm_nm'].str.split(expand=True).rename(columns={0:'시', 1:'구', 2:'동'})
], axis=1)
```

```python
hgd = hgd[hgd['시'] == '서울특별시'][['시', '구', '동', '행정동코드', 'geometry']]
```
